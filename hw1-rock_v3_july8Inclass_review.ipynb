```python
import requests
import pandas as pd
from lxml import etree
from tqdm import tqdm
```

```python
url = "https://rocktumbler.com/blog/rock-and-mineral-clubs/"
```

```python
headers = {
    "User-Agent": "Mozilla/5.0"
    }
     
resp = requests.get(url, headers=headers)

print("status:", resp.status_code)           # should print 200, not 403
```

```python
contents = resp.text
contents[:500]
```

#### Step 1 

```python
tree = etree.HTML(resp.text)
```

```python
xpath_tables = './/table[@class="font12"]'
tables = tree.xpath(xpath_tables)
len(tables) 
```

### Step 2 - work only with the first table as a template for checks

```python
table = tables[0]
```

```python
#view the HTML of one table as a check 
etree.tostring(table, encoding = 'unicode')
```

```python
table.xpath('//text()')
```

```python
table.xpath('.//tr')
#count how many rows it returns - compare against the website 
```

#### Step 3 - it all checks up for the first table, now start buidling variables to hold the links

```python
rows = table.xpath ('.//tr')
```

```python
row = rows[0]
```

```python
row.xpath('.//text()')
```

```python
xpath_url = './/a/@href'
row.xpath(xpath_url)
```

```python
#use normalize space to strip whitespaces, instead of a list
row.xpath(f'normalize-space({xpath_url})')
```

```python
xpath_club_name = './/a'
row.xpath(f'normalize-space({xpath_club_name})')
```

```python
xpath_city = './/td[@width = "40%"]'
row.xpath(f'normalize-space({xpath_city})')
```

```python
xpath_state_title = './/preceding-sibling::table//h3//text()'
table.xpath(f'normalize-space({xpath_state_title})')
```
